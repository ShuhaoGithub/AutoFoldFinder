```python
import numpy as np
from collections import Counter
space = np.genfromtxt('structure_matrix_exp3.csv',delimiter=',')
label = np.arange(250)
threshold = 0.8
for i in range(250):
    for j in range(i,250):
        if space[i,j] > threshold:
            label[i] = min(label[i],label[j])
            label[j] = min(label[i],label[j])
mask = np.unique(label)

len(Counter(label))
```
