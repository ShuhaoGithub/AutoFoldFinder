```python
import glob
import pandas as pd
exp = 104
count = len(glob.glob(f"../data/experiments/exp{exp}/output/final_seq_*_model.pdb"))
record = []
for i in range(count):
    with open(f"../data/experiments/exp{exp}/output/final_seq_{i}_model.txt",'r') as f:
        data = f.readlines()
        h = 0
        e = 0
        for j in range(28,len(data)):
            if data[j][16] == 'E':
                e += 1
            elif data[j][16] == 'H':
                h += 1
    record.append([i,e/(len(data)-28),h/(len(data)-28)])
df = pd.DataFrame(record, columns = ['file', 'sheet', 'helix'])
df.to_csv(f"../data/experiments/exp{exp}/exp{exp}_secondary_structure_ratio.csv", index = None)
```
